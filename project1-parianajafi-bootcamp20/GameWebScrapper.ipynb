```python
import requests
from bs4 import BeautifulSoup
```

```python
import re
```

```python
scrapped_data = pd.DataFrame(index = data['title'] ,
                             columns =['like','neutral', 'unlike', 'add','status',
                                       'Member_Rating', 'Reviews', 'Popularity', 'percent'])
scrapped_data
```

```python
for i,url in enumerate(data['freetogame_profile_url']):

    
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
    response = requests.get(url,headers=headers )
    results = BeautifulSoup(response.text, 'html.parser')

    first_elements = results.find_all("div", class_="count mb-n2 text-muted")
    first_elements = [i.text for i in first_elements]
    

    second_elements = results.find_all("div", class_="statRightHolder mb-2")
    second_elements = [i.text for i in second_elements]
    
    temp = re.findall(r'\d+', second_elements[0])
    res = list(map(int, temp))
    elements =[res[0], res[2], res[3]]
    
    ind = second_elements[0].index(str(res[0]))
    score = [second_elements[0][:ind].strip()]
    second_elements = score + elements
    
    
    percent= results.find_all("div",class_="progress-pie-chart")
    percent = [int(str(percent)[47:49])]
    
    new_row = first_elements + second_elements + percent
    index = data['title'][i]
    
    scrapped_data.loc[index] = new_row
    
    
    
```
