```python
from collections import Counter
import string
```

```python
def k_most_frequent_words(strings, k):
    combined_string = ' '.join(strings)
    stop_words = {"a", "an", "the", "and", "or", "with", "to", "of", "from", 
                  "is", "in", "on", "for", "at", "by", "as", "it", "this", "that", "these", "those"}
    words = combined_string.split()
    
    translator = str.maketrans('', '', string.punctuation)
    words = [word.translate(translator).lower() for word in words]
    
    filtered_words = [word for word in words if word.lower() not in stop_words]
    word_counts = Counter(filtered_words)
    most_common_words = word_counts.most_common(k)
    
    return most_common_words
```

```python
k = 100
print(k_most_frequent_words(full_data['short_description'], k))  
    
```

```python
genres = list(set(full_data['genre']))
genres
for genre in genres:
    k = 100
    strings = full_data[full_data['genre']==genre]
    strings = strings['short_description']
    print(genre, '\n', k_most_frequent_words(strings, k),'\n')  
```
